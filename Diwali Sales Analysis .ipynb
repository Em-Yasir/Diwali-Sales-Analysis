# Exploratory Data Analysis


## Gender

```python
sns.countplot(x='Gender', data = df)
```

```python
# to put the labels on 
ax = sns.countplot(x='Gender', data = df)

for bars in ax.containers:
        ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
sns.barplot(x='Gender',y='Amount' ,data = sales_gen)
```

#### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.

## Age

```python
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax = sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
        ax.bar_label(bars)
```

```python
#Total amount vs age group

sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Age Group')
sns.barplot(data = sales_age, x = 'Age Group', y = 'Amount')
```

#### From above Graphs we can see that the most of the buyers are of age group between 26-35 and Females

## State

```python
#Total amount of orders from top 10 states

df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
```

```python
orders_state =df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)

plt.figure(figsize=(16, 5))

#sns.set(rc={'figure.figsize': (16, 5)})
sns.barplot(data=orders_state,x='State',y='Orders')
```

```python
#Total amount spent from top 10 states

df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
```

```python
sales_state =df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)

plt.figure(figsize=(16, 5))

#sns.set(rc={'figure.figsize': (16, 5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

#### From above Graphs we can see that most of the orders and total amount spent are from Uttar Pardesh, Maharashtra, and Karnataka respectively 

## Marital Status

```python
ax = sns.countplot(x='Marital_Status', data = df)

for bars in ax.containers:
        ax.bar_label(bars)
```

```python
df.groupby(['Gender','Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sales_mar = df.groupby(['Gender','Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(x='Marital_Status',y='Amount' ,data = sales_mar,hue='Gender')
```

#### From above Graphs we can see that most of the buyers are married (Female) and they have high purchasing power

### Occupation

```python
ax = sns.countplot(x='Occupation', data = df)
plt.figure(figsize=(16, 5))
#sns.set(rc={'figure.figsize': (20, 5)})
for bars in ax.containers:
        ax.bar_label(bars)
```

```python
sales_occu =df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)

plt.figure(figsize=(16, 5))

#sns.set(rc={'figure.figsize': (16, 5)})
sns.barplot(data=sales_occu,x='Occupation',y='Amount')
```

#### From above Graphs we can see that most of the buyres are working in IT Sector, Healthcare and Aviation Sector

## Product Category

```python
ax = sns.countplot(x='Product_Category', data = df)
plt.figure(figsize=(19, 5))
#sns.set(rc={'figure.figsize': (20, 5)})
for bars in ax.containers:
        ax.bar_label(bars)
```

```python
sales_occu =df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)

plt.figure(figsize=(20, 5))

#sns.set(rc={'figure.figsize': (16, 5)})
sns.barplot(data=sales_occu,x='Product_Category',y='Amount')
```

### From above Graphs we can see that most of the product solds are from Food, Clothing and Electronics Category

```python
sales_occu =df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)

plt.figure(figsize=(20, 5))

#sns.set(rc={'figure.figsize': (16, 5)})
sns.barplot(data=sales_occu,x='Product_ID',y='Orders')
```

```python
# Top 10 most sold products same as above

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby(['Product_ID'])['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
